```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split

iris = datasets.load_iris()
features = iris.data
target = iris.target
```

```python
df = pd.DataFrame(data=iris.data, columns=iris.feature_names) 
df["target"] = iris.target
```

```python
"""
  Split the test data and train data
"""

X_train, X_test, y_train, y_test = train_test_split( features, target, test_size=0.1, random_state=42 )
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
"""
  implement k-means clustering
"""
predictor = KMeans(n_clusters = 3, random_state = 42) 
train_labels = predictor.fit_predict(X_train)
```

```python
"""
  Visualize the clustering
  Plot the identified clusters and compare with the answers
"""
fig, axes = plt.subplots(1, 2, figsize=(16,8))
axes[0].scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap='gist_rainbow', edgecolor='k', s=150)
axes[1].scatter(X_train[:, 0], X_train[:, 1], c=train_labels, cmap='jet', edgecolor='k', s=150)
axes[0].set_xlabel('Sepal length', fontsize=14)
axes[0].set_ylabel('Sepal width', fontsize=14)
axes[1].set_xlabel('Sepal length', fontsize=14)
axes[1].set_ylabel('Sepal width', fontsize=14) 
axes[0].tick_params(direction='in', length=10, width=5, colors='b',labelsize=20)
axes[1].tick_params(direction='in', length=10, width=5, colors='b',labelsize=20) 
axes[0].set_title('Actual', fontsize=14) 
axes[1].set_title('Clustered', fontsize=14)
```

```python
"""
  predict the labels
"""
labels = predictor.predict(X_test)
```

```python
"""
  Visualize the result
"""
fig, axes = plt.subplots(1, 2, figsize=(16,8))
axes[0].scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap='gist_rainbow',edgecolor='b', s=150)
axes[1].scatter(X_test[:, 0], X_test[:, 1], c=labels, cmap='jet', edgecolor='b', s=150)
axes[0].set_xlabel('Sepal length', fontsize=14)
axes[0].set_ylabel('Sepal width', fontsize=14)
axes[1].set_xlabel('Sepal length', fontsize=14)
axes[1].set_ylabel('Sepal width', fontsize=14)
axes[0].tick_params(direction='in', length=10, width=5, colors='b', labelsize=20)
axes[1].tick_params(direction='in', length=10, width=5, colors='b', labelsize=20) 
axes[0].set_title('Actual', fontsize=18) 
axes[1].set_title('Predicted', fontsize=18)
```
